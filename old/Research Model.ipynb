```python
#Code for Model implementation
#Author: Abilash Raghuram 

#1 = moral values
#2 = Propriety
#3 = Sanctity
#4 = Liberty
#5 = Care
#6 = Fairness
#7 = Authority
#8 = Loyality
#9 = Politeness

import numpy as np
import math as math
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt

concepts = []
h1 = []
h2 = []
h3 = []
h1.append([[5,6,7,8,9,3,4],[3],[4],[5,6,7,8,9]]) #H1

h2.append([[5,6,7,8,9,3],[5,6,7,8,9,4],[3,4],[5,6],[5,7],[5,8],[5,9],[5,3]
                ,[5,4],[6,7],[6,8],[6,9],[6,3],[6,4],[7,8],[7,9],[7,3],[7,4],[8,9],
                [8,3],[8,4],[9,3],[9,4],[3,4]])  #H2

h3.append([[5,6,7,8,9,3,4],[5,6,7],[5,6,8],[5,6,9],[5,6,3],[5,6,4],[5,7,8],[5,7,9],[5,7,3],
                 [5,7,4],[5,8,9],[5,8,3],[5,8,4],[5,9,3],[5,9,4],[5,3,4],[6,7,8],[6,7,9],[6,7,3],
                [6,7,4],[6,8,9],[6,8,3],[6,8,4],[6,9,3],[6,9,4],[6,3,4],[7,8,9],[7,8,3],[7,8,4]
                ,[7,9,3],[7,9,4],[7,3,4],[8,9,3],[8,9,4],[8,3,4],[9,3,4]]) #H3


concepts.append(h1)
concepts.append(h2)
concepts.append(h3)


    
def computeLiklihood(C , X):
    matches = 0 
    for h in range(0 , len(C)):
        for i in range (0, len(X)):
            if ((C[h] == X[i])):
                matches = matches + 1
    if (matches == len(X)):
        return (1/len(C))**len(X)
    else:
        return 0
    

#findHypLayer return the kth layer in which x in contained
#the minimal layer in which the union is found is considered in  case of conflictions 
def findHypLayer(x): 
    
    for i in range(len(h1)):    
        for j in range(len(h1[i])):
            if (x == h1[i][j]):
                return 1
    for i in range(len(h2)):    
        for j in range(len(h2[i])):
            if (x == h2[i][j]):
                return 2
    for i in range(len(h3)):    
        for j in range(len(h3[i])):
            if (x == h3[i][j]):
                return 3

    
    return 0

def computePrior(x):
    k  = findHypLayer(x)
    lamb = 0.5
    return (lamb/(1-lamb))** k
    #return ((lamb)**k)*((1-lamb))**(19-k)

def containsHypo(C , x):
    for i in range(len(C)):
        if (x == C[i]):
            return 1
    return 0
def containsHypoArr(C , x):
    for i in range(len(C)):
        if (x[0] == C[i]):
            return 1
    return 0






x = [7] #load 2 behaviour attributes(numbers from 3-9) as inputs here

results =[0,0,0,0,0,0,0]

 
normalizedSum = 0 
for output in range(3,10):
    for i in range(len(concepts)):    
            for j in range(len(concepts[i])):
                for k in range(len(concepts[i][j])):
                    #print('here hypo',concepts[i][j][k])
                    if(containsHypo(concepts[i][j][k],output) ==1):
                        normalizedSum =normalizedSum +( computePrior(concepts[i][j][k])*computeLiklihood(concepts[i][j][k], x))
print('sumOfProbabilities', normalizedSum)





probSum = 0
y=[3]
for i in range(len(concepts)):    
        for j in range(len(concepts[i])):
            for k in range(len(concepts[i][j])):
                #print('here hypo',concepts[i][j][k])
                if(containsHypoArr(concepts[i][j][k],y) ==1):
                    probSum =probSum + computePrior(concepts[i][j][k])*computeLiklihood(concepts[i][j][k], x)
print('prob for Sanctity is ', probSum/normalizedSum)
results[0]=  probSum/normalizedSum

probSum = 0
y=[4]
for i in range(len(concepts)):    
        for j in range(len(concepts[i])):
            for k in range(len(concepts[i][j])):
                #print('here hypo',concepts[i][j][k])
                if(containsHypoArr(concepts[i][j][k],y) ==1):
                    probSum =probSum + computePrior(concepts[i][j][k])*computeLiklihood(concepts[i][j][k], x)
print('prob for Liberty is ', probSum/normalizedSum)
results[1]=  probSum/normalizedSum

probSum = 0
y=[5]
for i in range(len(concepts)):    
        for j in range(len(concepts[i])):
            for k in range(len(concepts[i][j])):
                #print('here hypo',concepts[i][j][k])
                if(containsHypoArr(concepts[i][j][k],y) ==1):
                    probSum =probSum + computePrior(concepts[i][j][k])*computeLiklihood(concepts[i][j][k], x)
print('prob for Care is ', probSum/normalizedSum)
results[2]=  probSum/normalizedSum   

probSum = 0
y=[6]
for i in range(len(concepts)):    
        for j in range(len(concepts[i])):
            for k in range(len(concepts[i][j])):
                #print('here hypo',concepts[i][j][k])
                if(containsHypoArr(concepts[i][j][k],y) ==1):
                    probSum =probSum + computePrior(concepts[i][j][k])*computeLiklihood(concepts[i][j][k], x)
print('prob for Fairness is ', probSum/normalizedSum)
results[3]=  probSum/normalizedSum

probSum = 0
y=[7]
for i in range(len(concepts)):    
        for j in range(len(concepts[i])):
            for k in range(len(concepts[i][j])):
                #print('here hypo',concepts[i][j][k])
                if(containsHypoArr(concepts[i][j][k],y) ==1):
                    probSum =probSum + computePrior(concepts[i][j][k])*computeLiklihood(concepts[i][j][k], x)
print('prob for Authority is ', probSum/normalizedSum)
results[4]=  probSum/normalizedSum
probSum = 0

y=[8]
for i in range(len(concepts)):    
        for j in range(len(concepts[i])):
            for k in range(len(concepts[i][j])):
                #print('here hypo',concepts[i][j][k])
                if(containsHypoArr(concepts[i][j][k],y) ==1):
                    probSum =probSum + computePrior(concepts[i][j][k])*computeLiklihood(concepts[i][j][k], x)
print('prob for Loyality is ', probSum/normalizedSum)
results[5]=  probSum/normalizedSum

probSum = 0
y=[9]
for i in range(len(concepts)):    
        for j in range(len(concepts[i])):
            for k in range(len(concepts[i][j])):
                #print('here hypo',concepts[i][j][k])
                if(containsHypoArr(concepts[i][j][k],y) ==1):
                    probSum =probSum + computePrior(concepts[i][j][k])*computeLiklihood(concepts[i][j][k], x)
print('prob for Politnesss is ', probSum/normalizedSum)
results[6]=  probSum/normalizedSum

# probSum = 0
# y=[8]
# for i in range(len(concepts)):    
#         for j in range(len(concepts[i])):
#             for k in range(len(concepts[i][j])):
#                 #print('here hypo',concepts[i][j][k])
#                 if(containsHypoArr(concepts[i][j][k],y) ==1):
#                     probSum =probSum + computePrior(concepts[i][j][k])*computeLiklihood(concepts[i][j][k], x)
# print('prob for Loyality is ', probSum/normalizedSum)
# results[7]=  probSum/normalizedSum

# probSum = 0
# y=[9]
# for i in range(len(concepts)):    
#         for j in range(len(concepts[i])):
#             for k in range(len(concepts[i][j])):
#                 #print('here hypo',concepts[i][j][k])
#                 if(containsHypoArr(concepts[i][j][k],y) ==1):
#                     probSum =probSum + computePrior(concepts[i][j][k])*computeLiklihood(concepts[i][j][k], x)
# print('prob for Politeness is ', probSum/normalizedSum)
# results[8]=  probSum/normalizedSum


# #print("count is ", tConcepts)


objects = ('Sanctity', 'Liberty', 'Care', 'Fairness','Authority',
          'Loyality','Politeness')
y_pos = np.arange(len(objects))
plt.figure(figsize=(9, 5))
plt.bar(y_pos, results, align='center', alpha=1)
plt.xticks(y_pos, objects)
plt.ylabel('Probability')
plt.title('Moral Attributes')

plt.show()

```
