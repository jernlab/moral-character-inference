```python
import pandas as pd
filename = './SingleInferenceTask_data'
df = pd.read_csv(filename + '.csv')
df = df.drop(['created','saved','shown','answered','optional','shown_relative','answered_relative','displaycount','hidden'],axis=1)
print(df.shape)
df.head()
```

```python
# find the part of the survey that actually has the data (i.e. remove all the blank rows without answers)
df = df.loc[[not x for x in df['answer'].isnull().values]]
print(df.shape)
df.head()
```

```python
# getting rid of the condition/consent buttons, but keeping the prompts
df = df.loc[~df['item_name'].isin(['condition','consent','instructions','code','feedback','feedback_text'])]
df = df.loc[~df['type'].isin(['note'])]
print(df.shape)
df.head()
```

```python
df['item_name'].value_counts()[:10]
```

```python
df['unit_session_id'].value_counts()
```

```python
# Count number of incomplete responses
sessions = df['unit_session_id'].value_counts()
no_response = list(sessions.loc[sessions < 10].index)
print('Number of incomplete responses:',len(no_response))
```

```python
# Select out only the rows corresponding to completed surveys
df = df.loc[[item not in no_response for item in df['unit_session_id']]]
print(df.shape)
df.head()
```

```python
# Filter out attention check fails
fails = []
for session_id in df['unit_session_id'].unique():
    df_session = df.loc[df['unit_session_id'] == session_id]
    df_attention = df_session.loc[[x.startswith('attention_') for x in df_session['item_name']]]
    if True in [x != str(1) for x in df_attention['answer'].values]:
        df = df.loc[[x != session_id for x in df['unit_session_id']]]
        fails.append(session_id)
print(df.shape)
print(fails,'failed the attention check. Total failures:',len(fails))
```

```python
# Remove rows for attention check questions, to not skew the data
df = df.loc[[not x.startswith('attention') for x in df['item_name']]]
print(df.shape)
df.head()
```

```python
# Anonymize subjects
unique = df['unit_session_id'].unique()
participants = df['unit_session_id']
for i in range(len(unique)):
    participants = participants.replace(to_replace=unique[i], value=i)
df = df.drop(['unit_session_id','session'],axis=1)
df['subject'] = participants
df.head()
```

```python
df['subject'].value_counts()
# All counts should be 21 (i.e. 21 rows per subject - 3 conditions x 7 conclusions per condition)
```

```python
# add information about premise category and question category
df['premise'] = [s[:3] for s in df['item_name']]
df['y'] = [s[3:5] for s in df['item_name']]
df.head()
```

```python
# get condition number
df['condition'] = [int(x.replace('condition ==','')) for x in df['showif']]
df = df.drop(['showif'],axis=1)
df.head()
```

```python
df['y'].value_counts()
# These should all be the same, since each conclusion is shown exactly once for each premise. 
```

```python
prompts = pd.read_csv('one_example_prompts.csv')
prompts.head()
```

```python
# convert prompts into a dictionary
dct = {}
for index, row in prompts.iterrows():
    dct[row['condition']] = row['categories']
print(dct)
```

```python
# update the premise categories in the dataframe
for index, row in df.iterrows():
    if row['premise'] == 'pr1':
        df.at[index,'premise'] = dct[row['condition']]
df.head()
```

```python
# Write cleaned-up dataset to file
df.to_csv(filename + '_processed.csv',index=False)
```

```python
# Debugging purposes only; saving information about incomplete surveys
df_full = pd.read_csv(filename + '.csv')
df_incomplete = df_full.loc[[item in no_response for item in df_full['unit_session_id']]]
# Note: This ID is hard coded because it has a lot more responses than the others; we're only using this to look at some
df_incomplete.to_csv(filename+'_incomplete.csv',index=False)
print(df_incomplete.shape)
df_incomplete['showif'].value_counts()
```
