```python
import pandas as pd
one = pd.read_csv('formr_one_example.csv')
three = pd.read_csv('formr_three_example.csv')
one.head()
```

```python
# pull out only the prompt rows into its own dataframe
one = one.loc[['prompt_' in x for x in one['name']]]
one.head()
```

```python
# narrow down to only those dealing with propriety, and only the columns we care about
one = one.loc[[x == 'pr1' for x in one['block_order']]]
one = one[['name','block_order','showif','label']]
one.head()
```

```python
# creating condition column
one['condition'] = [int(x.replace('condition == ','')) for x in one['showif']]
one = one.drop(['showif'],axis=1)
one.head()
```

```python
three = three.loc[['prompt_' in x for x in three['name']]]
three.head()
```

```python
# narrow down to only those dealing with propriety, and only the columns we care about
three = three.loc[[x == 'pd3' or x == 'pr3' for x in three['block_order']]]
three = three[['name','block_order','showif','label']]
three.head()
```

```python
# creating condition column (a number from 1-10)
three['condition'] = [int(x.replace('condition == ','')) for x in three['showif']]
three = three.drop(['showif'],axis=1)
three.head()
```

```python
# write out to CSV; manually add the categories
one.to_csv('formr_one_example_prompts.csv')
three.to_csv('formr_three_example_prompts.csv')
```
